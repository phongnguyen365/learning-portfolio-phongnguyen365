## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

### Things to note

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Your code here
plt.plot(x, y_1, c = 'green', ls = 'dashed', lw = 2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph')
plt.grid(alpha = 0.8)
plt.legend(loc='upper left')
# No label in plt.plot()
```

## 1.2 Adding another plot

```python
plt.plot(x, y_1, label = 'Y1', c = 'green', ls = 'dashed', lw = 1.5)
plt.plot(x, y_2, label = 'Y2', c = 'purple', lw = 1.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X vs. Y')
plt.grid(alpha = 0.3)
plt.legend(loc = 'lower right')
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .2, 0.6

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])

#Math
x3 = x**3
log2_x = np.log2(x)

plt.plot(x, x3, label = 'x^2', c = 'blue', ls = 'dashed')
plt.plot(x, log2_x, label = 'log2(x)', c = 'green')
plt.grid(alpha = 0.25)
plt.legend()
```

## We need another axis!

```python

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha = .3, c = 'purple')
plt.ylim([-8, 8])
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha = .3, c = 'green')
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
